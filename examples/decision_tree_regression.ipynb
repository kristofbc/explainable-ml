# Regression Decision Tree

```python
# Create a random dataset
# http://scikit-learn.org/stable/auto_examples/tree/plot_tree_regression.html#sphx-glr-auto-examples-tree-plot-tree-regression-py
x = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(x).ravel()
y[::5] += 3 * (0.5 - np.random.rand(16))
```

```python
# Fit our model
clf_depth_2 = RegressionDecisionTree(max_depth=2)
clf_depth_5 = RegressionDecisionTree(max_depth=5)
clf_depth_2.train(x, y)
clf_depth_5.train(x, y)

# Predict from our train model
x_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_pred_depth_2 = clf_depth_2.predict(x_test)
y_pred_depth_5 = clf_depth_5.predict(x_test)

# Plot the results
# http://scikit-learn.org/stable/auto_examples/tree/plot_tree_regression.html#sphx-glr-auto-examples-tree-plot-tree-regression-py
plt.figure(figsize=(20, 10))
plt.scatter(x, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(x_test, y_pred_depth_2, color="cornflowerblue", label="max_depth=2", linewidth=2)
plt.plot(x_test, y_pred_depth_5, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
